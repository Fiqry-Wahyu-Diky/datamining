### **2. Entrophy**

**Diskritisasi berbasis entropi** (entropy based discretization) adalah salah satu jenis algoritma diskritisasi terawasi yang menggunakan mekanisme top- down. Tujuan dari algoritma ini adalah mendapatkan partisi yang mengandung baris data dari kelas yang sama sebanyak mungkin.

```python
#labels

labels = ['kecil','sedang','besar']
```

```python
#menambah sample data
sample = combine[['sepal.length','SL.EWidth']]
sample.describe()
```

fungsi menghitung banyakanya data berdasarkan kategori

```python
np.random.seed(0)
sample["categori"] = np.where(np.random.choice(2,sample.shape[0]) < 1,"kecil","sedang")
sample
```

Membuat fungsi GetOverollCategory (keseluruhan) yang digunakan untuk menghitung data keseluruhan yang nantinya digunakan untuk menghitung entropy

```python
def overalCategory(sample, labels, col, category):
  group = sample.groupby(category).count()
  return [group.loc[label, col] for label in labels]
```

fungsi untuk prosses info D

```python
def split(nilai, sample, labels, col, category):
  less_group = sample[sample[col] < nilai]
  greater_group = sample[sample[col] >= nilai]

  length_less_group = overalCategory(less_group, labels, col, category)
  length_greater_group = overalCategory(greater_group, labels, col, category)

  return (length_less_group, length_greater_group)
```

Membuat fungsi entropy untuk melakukan perhitungan entropy

```python
def entropy(d):
  return -(sum([i/sum(d) * log2(i/sum(d)) if i/sum(d) != 0 else 0 for i in d]))
```

Membuat fungsi info dan gain

```python
def info(d, sample):
  return sum([(sum(i) / sample.shape[0]) * entropy(i) for i in d])
```

### **2.1 Gain**
<p align="justify">
Setelah membagi dataset berdasarkan sebuah atribut kedalam subset yang lebih kecil, entropi dari data tersebut akan berubah. Perubahan entropi ini dapat digunakan untuk menentukan bagus tidaknya pembagian data yang telah dilakukan. Perubahan entropi ini disebut dengan information gain dalam Algoritme C4.5. Information gain ini diukur dengan menghitung selisih antara entropi dataset sebelum dan sesudah pembagian (splitting) dilakukan. Pembagian yang terbaik
akan menghasilkan entropi subset yang paling kecil, dengan demikian berdampak pada information gain yang terbesar. 
</p>

$$
\begin{align*}
\displaystyle 𝐺𝑎𝑖𝑛(𝐴) &= 𝐸𝑛𝑡𝑟𝑜𝑝𝑖(𝑆) − \sum_{i=1}^{k}\frac{\left| S \right|}{\left| S_{i} \right|} * 𝐸𝑛𝑡𝑟𝑜𝑝𝑖(𝑆_{i}) \\
Keterangan &: \\
𝑆 &: himpunan \ kasus \\
𝐴 &: atribut \\
k &: jumlah \ partisi \ atribut \ A \\
|𝑆𝑖| &: jumlah \ kasus \ pada \ partisi \ ke-i \\
|𝑆| &: jumlah \ kasus \ dalam \ S 
\end{align*}
$$

menghitung nilai gain

```python
def gain(inisial, new):
  return inisial - new
```

menghitung nilai entrophy overall

```python
d = overalCategory(sample, labels, 'sepal.length', 'SL.EWidth')
Einisial = entropy(d)
Einisial
```

```python
#mencari  split 4.5
az = split(4.5, sample, labels, 'sepal.length', 'SL.EWidth')
Enew1 = info(az, sample)
gain(Einisial, Enew1)
```

```python
#mencari split 5.5
az = split(5.5, sample, labels, 'sepal.length', 'SL.EWidth')
Enew2 = info(az, sample)
gain(Einisial, Enew2)
```

```python
#mencari split 6.5
az = split(6.5, sample, labels, 'sepal.length', 'SL.EWidth')
Enew3 = info(az, sample)
gain(Einisial, Enew3)
```
